```python
# PART 1

import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()                         # You initialise a figure

axes1 = fig.add_axes([0, 0.6, 0.4, 0.4])   # You can put as many sets of axes on the figure as you want
axes2 = fig.add_axes([0.6, 0.6, 0.4, 0.4]) # Position the axis using a list taking
axes3 = fig.add_axes([0, 0, 0.4, 0.4])     # [x-start, y-start, width, height]
axes4 = fig.add_axes([0.6, 0, 0.4, 0.4])

x = np.arange(0, 15)
y = np.random.randn(15)

axes1.set_xlabel('Observation')
axes1.set_ylabel('Value')
axes1.set_title('A Random Graph')
axes1.plot(x, y)

x2 = np.arange(0,10)
y2 = x2 ** 2

axes2.plot(x2, y2) # You plot on the axis individually, treating them as objects
axes3.plot(x2, x2**0.5)
axes4.plot(x2, x2*np.random.randint(0,10))
```

```python
# PART 2.1

# You can also do this automatically as below:
fig,axes = plt.subplots(nrows=2, ncols=2, dpi=500) # axes is an array of nrows by ncols


axes[0][0].plot(x, y) # You index it the same way

plt.tight_layout() # Makes your graphs more organised

#fig.savefig('HighResPlot.jpeg')
```

```python
# PART 2.2

fig, ax = plt.subplots(dpi=100)


ax.plot(x, x**2, label='x squared')
ax.plot(x, x**3, label='x cubed')

plt.legend()
```
